```python
import pandas as pd

# Asegúrate de que el nombre del archivo sea correcto
df = pd.read_csv("data.csv")

# Ver las primeras filas
df.head()
```

```python
# Convertir la columna 'Date' a formato datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')

# Crear una nueva columna para el mes y año (Periodo mensual)
df['Month'] = df['Date'].dt.to_period('M')

# Verificar que se haya creado correctamente
df[['Date', 'Month']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurar un estilo limpio
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Convertir la columna de fecha al tipo datetime si no lo está ya
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Agregar una columna de mes para análisis temporal
df['Month'] = df['Date'].dt.to_period('M')

# Gráfico de líneas: Ingresos totales por mes
monthly_sales = df.groupby('Month')['Total'].sum().reset_index()
monthly_sales['Month'] = monthly_sales['Month'].astype(str)

plt.figure()
sns.lineplot(data=monthly_sales, x='Month', y='Total', marker='o')
plt.title("Ingresos Totales por Mes")
plt.xlabel("Mes")
plt.ylabel("Total de Ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Product line', y='Total')
plt.title("Distribución de Ingresos por Línea de Producto")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Quantity', y='Total', hue='Product line')
plt.title("Relación entre Cantidad Vendida y Total de Venta")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='City', y='Total')
plt.title("Distribución de Ingresos por Ciudad")
plt.tight_layout()
plt.show()
```

```python
# Agrupar por mes y línea de producto
sales_by_product_month = df.groupby(['Month', 'Product line'])['Total'].sum().reset_index()
sales_by_product_month['Month'] = sales_by_product_month['Month'].astype(str)

# Crear gráfico
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_by_product_month, x='Month', y='Total', hue='Product line', marker='o')
plt.title("Ingresos Mensuales por Línea de Producto")
plt.xlabel("Mes")
plt.ylabel("Ingresos Totales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Tabla dinámica
pivot_table = df.pivot_table(values='Total', index='Customer type', columns='Payment', aggfunc='mean')

# Mapa de calor
plt.figure(figsize=(8, 5))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title("Promedio de Ingresos por Tipo de Cliente y Método de Pago")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='gross income', y='Rating', hue='Gender', style='Customer type')
plt.title("Relación entre Ingreso Bruto y Calificación del Servicio")
plt.tight_layout()
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

# Seleccionar columnas relevantes
df_parallel = df[['Product line', 'Unit price', 'Quantity', 'Total', 'gross income', 'Rating']].copy()

# Para no tener muchas líneas, tomamos una muestra representativa
df_sample = df_parallel.groupby('Product line').apply(lambda x: x.sample(20, random_state=1)).reset_index(drop=True)

# Normalizar para que todas las variables estén en la misma escala
for col in ['Unit price', 'Quantity', 'Total', 'gross income', 'Rating']:
    df_sample[col] = (df_sample[col] - df_sample[col].min()) / (df_sample[col].max() - df_sample[col].min())

# Gráfico
plt.figure(figsize=(12, 6))
parallel_coordinates(df_sample, 'Product line', colormap=plt.get_cmap("Set2"))
plt.title("Coordenadas Paralelas: Comparación de Productos en Múltiples Variables")
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Seleccionar variables numéricas
features = ['Unit price', 'Quantity', 'Total', 'gross income', 'Rating']
x = df[features]

# Escalamiento
x_scaled = StandardScaler().fit_transform(x)

# PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x_scaled)

# Convertir a DataFrame
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['Product line'] = df['Product line']

# Gráfico
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Product line')
plt.title("PCA: Visualización Multivariada Reducida a 2D")
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Crear gráfico 3D
fig = px.scatter_3d(df,
                    x='Unit price',
                    y='Quantity',
                    z='Total',
                    color='Product line',
                    title='Relación entre Precio Unitario, Cantidad y Total en 3D',
                    opacity=0.7)

fig.show()
```

```python
# Instalar Streamlit si no está instalado
!pip install streamlit

# Ahora puedes importar Streamlit y el resto de las bibliotecas
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Configuración inicial
st.set_page_config(page_title="Dashboard de Ventas", layout="wide")

# Título
st.title("📊 Dashboard Interactivo de Ventas de Tiendas de Conveniencia")

# Subir archivo
uploaded_file = st.file_uploader("Sube el archivo CSV con los datos de ventas", type=["csv"])
if uploaded_file is not None:
    df = pd.read_csv(uploaded_file)

    # Preprocesamiento
    df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce') # Added errors='coerce' for robust parsing
    df['Month'] = df['Date'].dt.to_period('M').astype(str)

    # Filtros
    st.sidebar.header("Filtros")
    city_filter = st.sidebar.multiselect("Ciudad", options=df['City'].unique(), default=df['City'].unique())
    product_filter = st.sidebar.multiselect("Línea de Producto", options=df['Product line'].unique(), default=df['Product line'].unique())
    customer_filter = st.sidebar.multiselect("Tipo de Cliente", options=df['Customer type'].unique(), default=df['Customer type'].unique())

    df_filtered = df[
        (df['City'].isin(city_filter)) &
        (df['Product line'].isin(product_filter)) &
        (df['Customer type'].isin(customer_filter))
    ]

    st.subheader("📈 Ingresos por Mes")
    # Asegurarse de que 'Month' esté ordenado correctamente para el gráfico de línea
    sales_by_month = df_filtered.groupby('Month')['Total'].sum().reset_index()
    sales_by_month['Month'] = pd.to_datetime(sales_by_month['Month']).dt.strftime('%Y-%m') # Convert to sortable format
    sales_by_month = sales_by_month.sort_values('Month') # Sort by month
    fig1 = px.line(sales_by_month, x='Month', y='Total', markers=True, title="Total de Ventas por Mes")
    st.plotly_chart(fig1, use_container_width=True)

    st.subheader("📦 Distribución de Ingresos por Categoría de Producto")
    # Matplotlib/Seaborn figure needs to be explicitly created and passed to st.pyplot
    fig2, ax = plt.subplots()
    sns.boxplot(data=df_filtered, x='Product line', y='Total', ax=ax)
    ax.set_title("Boxplot de Ingresos por Línea de Producto")
    ax.tick_params(axis='x', rotation=45)
    plt.tight_layout() # Ensure labels don't overlap
    st.pyplot(fig2)
    plt.close(fig2) # Close the figure to free up memory

    st.subheader("🧊 Visualización 3D: Precio, Cantidad y Total")
    fig3 = px.scatter_3d(df_filtered, x='Unit price', y='Quantity', z='Total', color='Product line',
                         title="Relación entre Precio Unitario, Cantidad y Total (3D)", opacity=0.7)
    st.plotly_chart(fig3, use_container_width=True)

    st.success(f"Datos cargados correctamente: {df_filtered.shape[0]} registros filtrados.")
else:
    st.info("Por favor, sube un archivo CSV para comenzar.")
```

```python
# --- CÓDIGO PARA dashboard_tarea_grupo_X.py ---
# (Este bloque NO se ejecuta directamente en Jupyter)

# Utiliza %%writefile magic command para guardar el contenido de la celda en un archivo .py
# Si el archivo ya existe, lo sobrescribirá.
%%writefile dashboard_tarea.py
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Configuración inicial
st.set_page_config(page_title="Dashboard de Ventas", layout="wide")

# Título
st.title("📊 Dashboard Interactivo de Ventas de Tiendas de Conveniencia")

# Subir archivo
# En Streamlit Cloud, puedes subir el archivo al repo. Para ejecutar local,
# puedes usar st.file_uploader o asegurar que el CSV está en el mismo dir.
# Simplificaremos para que funcione si el CSV está en el mismo directorio que dashboard_tarea.py
try:
    df = pd.read_csv("data.csv") # Assuming data.csv is in the same directory
except FileNotFoundError:
    st.error("Error: No se encontró el archivo 'data.csv'. Por favor, asegúrate de que esté en el mismo directorio que 'dashboard_tarea.py'.")
    st.stop() # Stop the app if file is not found

# Preprocesamiento
# Asegurar que la columna 'Date' existe antes de procesar
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')
    # Eliminar filas con fechas no válidas si errors='coerce' resultó en NaT
    df.dropna(subset=['Date'], inplace=True)
    if not df.empty: # Ensure DataFrame is not empty after dropping NaT
        df['Month'] = df['Date'].dt.to_period('M').astype(str)
    else:
        st.warning("No hay datos válidos de fecha después del preprocesamiento.")
        st.stop()
else:
     st.error("Error: La columna 'Date' no se encontró en el archivo CSV.")
     st.stop()


# Filtros
st.sidebar.header("Filtros")
# Usar try-except para manejar casos donde las columnas de filtro puedan faltar
try:
    city_filter = st.sidebar.multiselect("Ciudad", options=df['City'].unique(), default=df['City'].unique())
    product_filter = st.sidebar.multiselect("Línea de Producto", options=df['Product line'].unique(), default=df['Product line'].unique())
    customer_filter = st.sidebar.multiselect("Tipo de Cliente", options=df['Customer type'].unique(), default=df['Customer type'].unique())

    df_filtered = df[
        (df['City'].isin(city_filter)) &
        (df['Product line'].isin(product_filter)) &
        (df['Customer type'].isin(customer_filter))
    ]
except KeyError as e:
    st.error(f"Error: Columna de filtro no encontrada: {e}. Asegúrate de que tu CSV contiene las columnas 'City', 'Product line', y 'Customer type'.")
    st.stop()


# --- Visualizaciones (Asegúrate de que las columnas existan y los datos sean apropiados) ---

if not df_filtered.empty:

    # Gráfico 1: Ingresos por Mes (Plotly)
    if 'Month' in df_filtered.columns and 'Total' in df_filtered.columns:
        st.subheader("📈 Ingresos por Mes")
        sales_by_month = df_filtered.groupby('Month')['Total'].sum().reset_index()
        # Convertir a datetime para ordenar correctamente
        sales_by_month['Month_dt'] = pd.to_datetime(sales_by_month['Month'])
        sales_by_month = sales_by_month.sort_values('Month_dt')
        # Usar la columna original 'Month' (como string) para el eje X si es necesario o formatear
        sales_by_month['Month_str'] = sales_by_month['Month_dt'].dt.strftime('%Y-%m')
        fig1 = px.line(sales_by_month, x='Month_str', y='Total', markers=True, title="Total de Ventas por Mes")
        fig1.update_layout(xaxis_title="Mes", yaxis_title="Total de Ventas")
        st.plotly_chart(fig1, use_container_width=True)
    else:
         st.warning("Saltando gráfico de Ingresos por Mes: Columnas 'Month' o 'Total' no encontradas.")


    # Gráfico 2: Distribución de Ingresos por Categoría de Producto (Matplotlib/Seaborn)
    if 'Product line' in df_filtered.columns and 'Total' in df_filtered.columns:
        st.subheader("📦 Distribución de Ingresos por Categoría de Producto")
        fig2, ax = plt.subplots(figsize=(10, 6)) # Define figure size
        sns.boxplot(data=df_filtered, x='Product line', y='Total', ax=ax)
        ax.set_title("Boxplot de Ingresos por Línea de Producto")
        ax.set_xlabel("Línea de Producto")
        ax.set_ylabel("Total de Ventas")
        ax.tick_params(axis='x', rotation=45)
        plt.tight_layout() # Ensure labels don't overlap
        st.pyplot(fig2)
        plt.close(fig2) # Close the figure to free up memory
    else:
         st.warning("Saltando gráfico de Distribución de Ingresos por Categoría: Columnas 'Product line' o 'Total' no encontradas.")


    # Gráfico 3: Visualización 3D (Plotly Express)
    if all(col in df_filtered.columns for col in ['Unit price', 'Quantity', 'Total', 'Product line']):
        st.subheader("🧊 Visualización 3D: Precio, Cantidad y Total")
        fig3 = px.scatter_3d(df_filtered, x='Unit price', y='Quantity', z='Total', color='Product line',
                            title="Relación entre Precio Unitario, Cantidad y Total (3D)", opacity=0.7)
        fig3.update_layout(scene = dict(
                            xaxis_title='Precio Unitario',
                            yaxis_title='Cantidad',
                            zaxis_title='Total'))
        st.plotly_chart(fig3, use_container_width=True)
    else:
        st.warning("Saltando Visualización 3D: Columnas 'Unit price', 'Quantity', 'Total' o 'Product line' no encontradas.")


    st.success(f"Datos cargados y filtrados correctamente: {df_filtered.shape[0]} registros mostrados.")

else:
    st.info("No hay datos para mostrar con los filtros seleccionados.")

# %%
```
